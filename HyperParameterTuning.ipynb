```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/Iris.csv")
df.head()
```

```python
df["Species"].value_counts()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns = ["Species"], axis = 1)
y = df["Species"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# KNN model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
knn_model_pred = knn_model.predict(X_test)
knn_model_score = knn_model.score(X_test, y_test)
knn_model_score
```

```python
# SVC model
from sklearn.svm import SVC
svc_model = SVC(gamma = "auto")
svc_model.fit(X_train, y_train)
svc_model_pred = svc_model.predict(X_test)
svc_model_score = svc_model.score(X_test, y_test)
svc_model_score
```

```python
# Grid Search CV
from sklearn.model_selection import GridSearchCV
svm_classifier = GridSearchCV((svc_model),{
    "C" : [1, 10, 20, 30],
    "kernel": ["linear", "rbf"]
}, cv = 5, return_train_score = False)
```

```python
svm_classifier.fit(X, y)
```

```python
svm_classifier.cv_results_
```

```python
results = pd.DataFrame(svm_classifier.cv_results_)
results.head()
```

```python
results.sort_values(by = ["mean_test_score"], ascending = False)[["param_C", "param_kernel", "mean_test_score"]]
```

```python
knn_classifier = GridSearchCV((knn_model), {
    "n_neighbors": [3, 5, 7, 9, 13],
    "weights": ["uniform", "distance"],
    "algorithm": ["auto", "ball_tree", "kd_tree", "brute"]
}, cv = 5, return_train_score = False)
```

```python
knn_classifier.fit(X, y)
```

```python
knn_classifier.cv_results_
```

```python
knn_results = pd.DataFrame(knn_classifier.cv_results_)
```

```python
knn_results
```

```python
knn_results.sort_values(by = "mean_test_score", ascending = False)[["param_algorithm", "param_n_neighbors", "param_weights", "mean_test_score"]]
```

```python
# Random Search CV
from sklearn.model_selection import RandomizedSearchCV
```

```python
knn_random_classifier = RandomizedSearchCV((knn_model), {
    "n_neighbors": [3, 5, 7, 9, 13],
    "weights": ["uniform", "distance"],
    "algorithm": ["auto", "ball_tree", "kd_tree", "brute"]
},n_iter = 10, cv = 5, return_train_score = False)
```

```python
knn_random_classifier.fit(X, y)
```

```python
knn_random_classifier.cv_results_
```

```python
knn_random_result = pd.DataFrame(knn_random_classifier.cv_results_)
knn_random_result
```

```python
knn_random_result.sort_values(by = "mean_test_score", ascending = False)[["param_algorithm", "param_n_neighbors", "param_weights", "mean_test_score"]]
```
