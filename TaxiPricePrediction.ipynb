# Data cleaning and preprocessing

```python
final_df = df.copy()
```

```python
final_df.head()
```

```python
final_df = final_df.drop(columns = ['Passenger_Count'], axis = 1)
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
# remove duplicates
final_df = final_df.drop_duplicates()
```

```python
final_df.shape
```

```python
final_df.isnull().sum()
```

```python
final_df.columns
```

```python
final_df.info()
```

```python
numeric_cols
```

```python
final_df[numeric_cols] = final_df[numeric_cols].fillna(final_df[numeric_cols].mean())
```

```python
final_df.isnull().sum()
```

```python
categorical_cols = ['Time_of_Day', 'Day_of_Week', 'Traffic_Conditions', 'Weather']
```

```python
final_df[categorical_cols] = final_df[categorical_cols].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
final_df.isnull().sum()
```

```python
final_df['Weather'].value_counts()
```

```python
final_df['Traffic_Conditions'].value_counts()
```

```python
final_df['Day_of_Week'].value_counts()
```

```python
final_df['Time_of_Day'].value_counts()
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
encoding_cols = ['Time_of_Day', 'Day_of_Week', 'Weather']
```

```python
final_df = pd.get_dummies(final_df, columns = encoding_cols)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
final_df['Traffic_Conditions'] = encoder.fit_transform(final_df['Traffic_Conditions'])
```

```python
final_df.head()
```

```python
final_df = final_df.astype(int)
```

```python
final_df.head()
```

```python
final_df.shape
```

# Feature scaling

```python
final_df.columns
```

```python
final_df.info()
```

```python
final_df.head()
```

```python
scaling_cols = ['Trip_Distance_km', 'Base_Fare', 'Trip_Duration_Minutes' ]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
final_df[scaling_cols] = scaler.fit_transform(final_df[scaling_cols])
```

```python
final_df.head()
```

# Split Dataset

```python
X = final_df.drop(columns = ['Trip_Price'], axis = 1)
y = final_df['Trip_Price']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

# Model evaluation

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_pred)
print(f'R2 score: {r2}')
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1-((1-r2)*n-1)/(n-p-1)
print(f'Adjusted R2 score: {adjusted_r2}')
```
