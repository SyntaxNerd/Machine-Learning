# Data cleaning and preprocessing

```python
final_df = df.copy()
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df.shape
```

```python
final_df = final_df.drop_duplicates()
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
final_df['Stage_fear'].value_counts()
```

```python
final_df['Drained_after_socializing'].value_counts()
```

```python
final_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
final_df['Stage_fear'] = encoder.fit_transform(final_df['Stage_fear'])
final_df['Drained_after_socializing'] = encoder.fit_transform(final_df['Drained_after_socializing'])
final_df['Personality'] = encoder.fit_transform(final_df['Personality'])
```

```python
final_df.head()
```

# Scaling

```python
X = final_df.drop(columns = ['Personality'], axis = 1)
y = final_df['Personality']
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_cols = ['Time_spent_Alone', 'Social_event_attendance', 'Going_outside', 'Friends_circle_size', 'Post_frequency']
```

```python
X[scaled_cols] = scaler.fit_transform(X[scaled_cols])
```

```python
X.head()
```

# Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

# Model train

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
log_model = LogisticRegression()
knn_model = KNeighborsClassifier(n_neighbors = 5)
```

```python
log_model.fit(X_train, y_train)
knn_model.fit(X_train, y_train)
```

```python
log_pred = log_model.predict(X_test)
knn_pred = knn_model.predict(X_test)
```

# Model Evaluate

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, log_pred))
```

```python
print(classification_report(y_test, knn_pred))
```

```python
confusion_matrix(y_test, log_pred)
```

```python
confusion_matrix(y_test, knn_pred)
```
