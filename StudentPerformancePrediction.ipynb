# Data cleaning and preprocessing

```python
df_cleaned = df.copy()
```

```python
df_cleaned.head()
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned.shape
```

```python
df_cleaned = df_cleaned.drop_duplicates()
```

```python
df_cleaned.shape
```

```python
df_cleaned.head()
```

```python
df_cleaned['Extracurricular Activities'].value_counts()
```

```python
# encoding
df_cleaned = pd.get_dummies(df_cleaned, columns = ['Extracurricular Activities'])
```

```python
df_cleaned = df_cleaned.astype(int)
```

```python
df_cleaned.head()
```

# Feature eng and scaling

```python
df_cleaned.head()
```

```python
cols = ['Hours Studied', 'Previous Scores', 'Sleep Hours', 'Sample Question Papers Practiced']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_cleaned[cols] = scaler.fit_transform(df_cleaned[cols])
```

```python
df_cleaned.head()
```

```python
sns.heatmap(df_cleaned.corr(numeric_only = True), annot = True)
```

# Split dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_cleaned.drop(columns = ['Performance Index'], axis = 1)
y = df_cleaned['Performance Index']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

# Model training

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

# Model evaluation

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_pred)
print(f'R2 score: {r2}')
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1-((1-r2)*n-1)/(n-p-1)
print(f'Adjusted R2 score: {adjusted_r2}')
```
